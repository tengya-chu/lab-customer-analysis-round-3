```python
import pandas as pd
import numpy as np
data = pd.read_csv('./files_for_lab/csv_files/marketing_customer_analysis.csv')
```

```python
# Show DataFrame info.
from IPython.display import display
print ("data:")
display(data)
```

```python
# Describe DataFrame.
data.describe()
```

```python
# Show a plot of the total number of responses.
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
data['Response'].value_counts().plot(kind='bar')
plt.show()
```

```python
# Show a plot of the response rate by the sales channel.
sns.histplot(binwidth=0.5, x="Response", hue="Sales Channel", data=data, stat="count", multiple="stack")
```

```python
# Show a plot of the response rate by the total claim amount.
sns.boxplot(x = 'Response',y='Total Claim Amount', data=data)
plt.show()
```

```python
# Show a plot of the response rate by income.
sns.boxplot(x = 'Response',y='Income', data=data)
plt.show()
```
